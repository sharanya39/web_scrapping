```python
import cv2
import torch
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO
from deepface import DeepFace
from PIL import Image
```

```python
image_path = "/content/vijay_271.jpg"  # Change this to your image file
img = cv2.imread(image_path)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
# Analyze faces in the image
analysis = DeepFace.analyze(img_path=image_path, actions=['gender', 'emotion'])

for i, person in enumerate(analysis):
    gender = person['dominant_gender']
    emotion = person['dominant_emotion']
    print(f"Person {i+1}: Gender - {gender}, Emotion - {emotion}")
```
