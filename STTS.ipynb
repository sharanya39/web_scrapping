```python
# Step 2 - Import Required Libraries
import whisper
from transformers import pipeline
from gtts import gTTS
from google.colab import files
from IPython.display import Audio, display  # To play the audio
```

```python
# Step 3 - Upload the .wav File to Colab
uploaded = files.upload()
```

```python
# Assume the uploaded file is the question audio
input_wav = list(uploaded.keys())[0]  # Take the first uploaded file
```

```python
# Step 4 - Speech to Text (STT) using Whisper
def speech_to_text(wav_file):
    model = whisper.load_model("small")
    result = model.transcribe(wav_file)
    return result['text']

# Step 5 - Text Response Generation using Hugging Face (flan-t5-small)
def generate_answer(question_text):
    # Use a small and fast text generation model
    #model = "google/flan-t5-small"
    #qa_pipeline = pipeline("text2text-generation", model=model)
    qa_pipeline = pipeline("text2text-generation", model="google/flan-t5-large")
    # Original: qa_pipeline = pipeline("text-generation", model="mistralai/Mistral-7B-Instruct")
    #qa_pipeline = pipeline("text-generation", model="mistralai/Mistral-7B-Instruct-v0.1") # Added a version of the model



    # Add a prefix for question answering tasks
    input_prompt = f"Question: {question_text}\nAnswer:"
    print(f"Detected Question: {question_text}")

    # Generate answer
    response = qa_pipeline(input_prompt, max_length=100)
    return response[0]['generated_text']

# Step 6 - Text to Speech (TTS) using gTTS
def text_to_speech(answer_text, output_file="answer.wav"):
    tts = gTTS(text=answer_text, lang='en')
    tts.save(output_file)

# Step 7 - Full Pipeline
def main(input_wav):
    print(f"Processing file: {input_wav}")

    # Speech to Text
    print("Converting speech to text...")
    question_text = speech_to_text(input_wav)
    print(f"Detected Question: {question_text}")

    # Text Generation (AI Response)
    print("Generating answer using AI...")
    answer_text = generate_answer(question_text)
    print(f"Generated Answer: {answer_text}")

    # Text to Speech
    print("Converting answer text to speech...")
    text_to_speech(answer_text)

    # Play audio directly in Colab
    print("Playing generated answer...")
    display(Audio("answer.wav"))

    # Optionally download the file (uncomment if you want this feature)
    files.download("answer.wav")

# Step 8 - Run Everything
main(input_wav)
```
